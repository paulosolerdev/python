# 3. Métodos

```python
class Livro:

    def __init__(self, titulo, autor, ano):

        self.titulo = titulo
        self.autor = autor
        self.ano = ano

    def descricao(self):

        return f"'{self.titulo}' por {self.autor}, publicado em {self.ano}."
    
meu_livro = Livro("1984", "Geoge Orwell", 1949)

print(meu_livro.descricao())
```

```python
class Carro:

    def __init__(self, marca, modelo, cor):

        self.marca = marca
        self.modelo = modelo
        self.cor = cor
        self.velocidade = 0

    def acelerar(self):

        self.velocidade += 10

        print(f'Velocidade atual: {self.velocidade} km/h')

    def frear(self):

        if self.velocidade > 0:

            self.velocidade -= 10

            print(f'Velocidade atual: {self.velocidade} km/h')

        else:

            print('O carro está parado')
            
    def exibir_info(self):

        print(f'Marca: {self.marca}, Modelo: {self.modelo}, Cor: {self.cor}')

lista_carros = []

while True:

    print('\n--- Menu ---')
    print('1. Adicionar novo carro')
    print('2. Exibir informações dos carros')
    print('3. Acelerar um carro')
    print('4. Frear um carro')
    print('5. Sair')

    opcao = input('Escolha uma opção: ')
    
    if opcao == '1':

        marca = input('Digite a marca do carro: ')
        modelo = input('Digite o modelo do carro: ')
        cor = input('Digite a cor do carro: ')

        carro = Carro(marca, modelo, cor)

        lista_carros.append(carro)

    elif opcao == '2':

        if lista_carros:

            for carro in lista_carros:

                carro.exibir_info()

        else:
            
            print('Nenhum carro cadastrado')

    elif opcao == '3':

        modelo = input('Digite o modelo do carro que você deseja acelerar:  ')

        for carro in lista_carros:

            if carro.modelo == modelo:
                
                carro.acelerar()

                break

            else:
                
                print(f'Carro {modelo} não encontrado')

    elif opcao == '4':

        modelo = input('Digite o modelo do carro que você deseja frear: ')

        for carro in lista_carros:

            if carro.modelo == modelo:
                
                carro.frear()

                break
                
    elif opcao == '5':

        print('Saindo do programa...')

        break

    else:

        print('Opção inválida')
```

## 2. Conceitos Básicos

    Atributos: Definindo e acessando atributos de uma classe.

```python
# Exemplo Prático: Gerenciamento de Jogadores de Futebol

# Descrição:
# Imagine que você está desenvolvendo um software simples para gerenciar
# jogadores em um time de futebol. Cada jogador tem um nome, posição, 
# número da camisa e quantidade de gols marcados na temporada.

# 1. Definição da Classe Jogador:

class Jogador:

    def __init__(self, nome, posicao, numero_camisa, gols_marcados=0):

        self.nome = nome
        self.posicao = posicao
        self.numero_camisa = numero_camisa
        self.gols_marcados = gols_marcados


# 2. Instanciando e Acessando Atributos:

# Instanciando jogadores

jogador1 = Jogador('Cristiano Ronaldo', 'Atacante', 7)

jogador2 = Jogador('Lionel Messi', 'Atacante', 10)

jogador3 = Jogador('Neymar Jr.', 'Atacante', 10)

# Acessando atributos

print(f'{jogador1.nome} é um {jogador1.posicao} e usa a camisa número {jogador1.numero_camisa}')

print(f'{jogador2.nome} é um {jogador2.posicao} e usa a camisa número {jogador2.numero_camisa}')

print(f'{jogador3.nome} é um {jogador3.posicao} e usa a camisa número {jogador3.numero_camisa}')
```
