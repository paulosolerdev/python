```python
"""
Operações Básicas com Dicionários
    Adicionando itens
    Atualizando itens
    Removendo itens (del, pop(), popitem())
    Copiando dicionários (copy(), dict())
"""

# Operações Básicas com Dicionários

produto = {
    'id':12345,
    'nome':'Camisa Polo',
    'cor': 'Vermelho',
    'preco': 49.90,
    'estoque': 100
}

# 1. Adicionando itens

produto['marca'] = 'FashionBrand'

produto['desconto'] = 10

print(f'Após adicionar itens: {produto}')


# 2. Atualizando itens

produto['preco'] = 59.90
produto['desconto'] = 15

print(f'\nApós atualizar itens: {produto}')

print()

# 3. Removendo itens (del, pop(), popitem())

del produto['desconto']

print(produto)

cor_removida = produto.pop('cor')
print(f'\nCor removida: {cor_removida}')

item_removido = produto.popitem()

print(f'\nApós atualizar deletar itens: {produto}')

# 4. Copiando dicionários (copy(), dict())

# Método copy()
produto_copia_1 = produto.copy()

# Método dict()
produto_copia_2 = dict(produto)

print(f'\nCópias do produto: ')
print(f'Cópia 1: {produto_copia_1}')
print(f'Cópia 2: {produto_copia_2}')
```

```python
"""
Métodos de Dicionários
    Keys(), values(), e items()
    clear()
    setdefault()
    update()
    fromkeys()
"""

# Métodos de Dicionários

livro = {
    
    'titulo':'O Pequeno Príncipe',
    'autor':'Antoine de Saint-Exupéry',
    'ano':1943,
    'editora':'Reynal & Hitchcock',
    'preco':20.5
}

# 1. keys(), values(), e items()
print(f'\nkeys(), values(), e items()')
print(f'Chaves do dicionário: {list(livro.keys())}')
print(f'Valores do dicionário: {list(livro.values())}')
print(f'Itens do dicionário: {list(livro.items())}')

# 2. clear()
print(f'\n2. clear()')
copia_livro = livro.copy()
copia_livro.clear()
print(f'\nDicionários após o clear(): {copia_livro}')

# 3. setdefault()
print(f'\n3. setdefault()')
isbn = livro.setdefault('ISBN', '978-3-16-148410-0')
print(f'\nISBN {isbn}')
print(f'Dicionário após setdefault(): {livro}')

# 4. update()
print(f'\n4. update()')

atualizacoes = {
    'preco': 18.50,
    'formato': 'Capa dura'
}

livro.update(atualizacoes)

print(f'\nDicionário após o update(): {livro}')

# 5. fromkeys()
print(f'\n5. fromkeys()')
chaves = ['Titulo', 'Autor', 'sinopse']
novo_livro = dict.fromkeys(chaves, 'Desconhecido')
print(f'\nDicionário criado com fromkeys(): {novo_livro}')
```

```python
"""
Exercício sobre Métodos de Dicionários

Objetivo: Utilizando um dicionário que contém informações sobre um filme,
pratique o uso dos métodos de dicionários.

Dicionário fornecido:

filme = {
    'titulo': 'Inception',
    'diretor': 'Christopher Nolan',
    'ano': 2010,
    'genero': 'Ficção Científica'
}

Instruções:

    1. keys(), values() e items()
        - Imprima todas as chaves do dicionário filme.
        - Imprima todos os valores do dicionário filme.
        - Imprima todos os pares chave-valor do dicionário filme em 
        formato de lista de tuplas.
        
    2. clear()
        - Faça uma cópia do dicionário filme e limpe todos os elementos
        dessa cópia. Verifique se a cópia foi realmente esvaziada.
        
    3. setdefault()
        - Verifique se a chave 'elenco' existe no dicionário filme. Se
        não existir, adicione-a com o valor ['Leonardo DiCaprio', 'Ellen
        Page'].
    
    4. update()
        - Atualize o dicionário filme com as seguintes informações:
        duração de 148 minutos e idioma como 'Inglês'.
        
    
"""
filme = {
    'titulo': 'Inception',
    'diretor': 'Christopher Nolan',
    'ano': 2010,
    'genero': 'Ficção Científica'
}

# 1. keys(), values() e items()
print('\n1. keys(), values(), e items()')
print(f'Chaves: {list(filme.keys())}')
print(f'Valores: {list(filme.values())}')
print(f'Itens: {list(filme.items())}')

# 2. clear()
print(f'\n2.clear()')
copia_filme = filme.copy()
copia_filme.clear()
print(f'Dicionário após clear(): {copia_filme}')

# 3. setdefault()
print(f'\n3. setdefault()')
elenco = filme.setdefault('elenco', ['Leonardo DiCaprio', 'Ellen Page'])
print(f'\nElenco: {elenco}')
print(f'Dicionário após o setdefault(): {filme}')

# 4. update()
print(f'\n4. update()')
informacoes_adicionais = {
    'duracao': 148,
    'idioma': 'Inglês'
}
filme.update(informacoes_adicionais)
print(f'Dicionário após update(): {filme}')

# 5. fromkeys()
print(f'\n5. fromkeys()')
chaves = ['nome', 'idade', 'ocupacao']
perfil = dict.fromkeys(chaves, 'Desconhecido')
print(f'\nDicionário criado com fromkeys(): {perfil}')
```

```python
"""
Iterando Sobre Dicionários
    Usando loops for
    Iterando sobre chaves, valores e itens
"""

# Iterando Sobre Dicionários

# Exemplo Prático:

notas = {
    'Matemática': 8.5,
    'Português': 9.0,
    'História': 7.5,
    'Geografia': 8.0,
    'Química': 9.5
}

# 1. Iteração sobre as chaves do dicionário (neste caso, são as matérias)

print(f'Matérias cursadas pelo aluno: ')

for materia in notas:
    
    print(materia)

print(f'\nMatérias (usando .keys()): ')

for materia in notas.keys():
    
    print(materia)

# 2. Iterando apenas sobre os valores (notas)

print(f'\nNotas do aluno: ')

total = 0

for nota in notas.values():
    
    print(nota)
    
    total += nota
    
media = total / len(notas)

print(f'\nMédia das notas: {media}')

# 3. Iterando sobre chaves e valores simultaneamente (items)

print(f'\nRelatório de notas: ')

for materia, nota in notas.items():
    
    print(f'{materia}: {nota}')

```

```python
"""
Exercício sobre Iterando Sobre Dicionários

Objetivo: Você tem um dicionário que representa o número de livros vendidos
em uma livraria em diferentes meses. Sua tarefa é iterar sobre esse 
dicionário para realizar diferentes análises.

Dicionário fornecido:

vendas = {
    'Janeiro': 120,
    'Fevereiro': 150,
    'Março': 80,
    'Abril': 190,
    'Maio': 210
}

Instruções:

    1. Iterando sobre as chaves (meses):
        - Imprima todos os meses em que a livraria registrou vendas.
        
    2. Iterando apenas sobre os valores (números de livros vendidos):
        - Calcule e imprima a venda total nos 5 meses.
        - Determine e imprima o mês com as vendas mais baixas.
        
    3. Iterando sobre chaves e valores simultaneamente (items):
        - Para cada mês, imprima uma mensagem no seguinte formato:
        'Em [mês], [número] livros foram vendidos'.
"""

vendas = {
    'Janeiro': 120,
    'Fevereiro': 150,
    'Março': 80,
    'Abril': 190,
    'Maio': 210
}

1. 
print(f'Meses de vendas registradas: ')
for venda in vendas:
    
    print(f'{venda}')

2.
total = 0
vendas_mais_baixas = float('inf')
mes_vendas_mais_baixas = ''

print(f'\n')

for venda in vendas.values():
    
    total += venda
    
    if venda < vendas_mais_baixas:
        
        vendas_mais_baixas = venda
        
for mes, venda in vendas.items():
    
    if venda == vendas_mais_baixas:
        
        mes_vendas_mais_baixas = mes
        
        break
        
print(f'\nTotal de livros vendidos: {total}')

print(f'Mês com as vendas mais baixas: {mes_vendas_mais_baixas} com {vendas_mais_baixas} vendidos.')


# 3. Iterando sobre chaves e valores simultaneamente (items):

print(f'\n3. Iterando sobre chaves e valores simultaneamente')

for mes, venda in vendas.items():
    
    print(f'Em {mes}, {venda} livros foram vendidos.')
```

```python
"""
Compreensão de Dicionários (Dictionary Comprehension)
    Sintaxe básica
    Exemplos práticos
"""

# A compreensão de dicionários é uma maneira concisa de criar dicionários.
# Assim como a compreensão de listas em Python, ela oferece uma forma
# mais legível e muitas vezes mais rápida de criar dicionários do que
# usar loops for tradicionais.

# Sintaxe básica:

# {chave: valor for item in iterável}

# Exemplos Práticos:

# 1. Criar um dicionário com quadrados de números:

# Suponha que queremos criar um dicionário onde as chaves são números
# de 1 a 5 e os valores são seus quadrados.

quadrados = {x: x**2 for x in range(1, 6)}
print(quadrados)

# usando um loop for tradicional

quadrados = {}

for  x in range(1, 6):
    
    quadrados[x] = x**2
    
print(quadrados)

# 2. Converter chaves em valores e valores em chaves:

original = {'a': 1, 'b': 2, 'c': 3}
invertido = {valor: chave for chave, valor in original.items()}
print(invertido)

original = {'a': 1, 'b': 2, 'c': 3}
invertido = {}

for chave, valor in original.items():
    
    invertido[valor] = chave
    
print(invertido)

# 3. Filtrando itens de um dicionário:

precos = {
    'laptop': 1000,
    'mouse': 25,
    'monitor': 200,
    'teclado': 30,
    'cabo hdmi': 10
}

caros = {produto: preco for produto, preco in precos.items() if preco > 50}

print(caros)

caros = {}

for produto, preco in precos.items():
    
    if preco > 50:
        
        caros[produto] = preco
        
print(caros)

# 4. Dicionário com palavras e seus comprimentos:

palavras = {'Python', 'compreensão', 'dicionário'}
comprimentos = {palavra: len(palavra) for palavra in palavras}
print(comprimentos)

# Usando um loop for tradicional, você faria o seguinte:

palavras = ['Python', 'compreensão', 'dicionário']

comprimentos = {}

for palavra in palavras:
    
    comprimentos[palavra] = len(palavra)
    
print(comprimentos)
```

```python
"""
Exercício: Compreensão de Dicionários (Dicionary Comprehension)

Objetivo: Familiarizar-se com a sintaxe básica da compreensão de 
dicionários em Python.

Instruções:

Dada a seguinte lista de números:

numeros = [1, 2, 3, 4, 5]

a) Use a compreensão de dicionário para criar um novo dicionário onde
as chaves são os números da lista e os valores são os quadrados desses
números.

Resultado esperado:

{1: 1, 2: 4, 3: 9, 4: 16, 5: 25}

Para elevar um número ao quadrado, use o operador **. Por exemplo, 3**2
resulta em 9.

"""

numeros = [1, 2, 3, 4, 5]

dicionario = {x: x**2 for x in numeros}

print(dicionario)

quadrados = {}

for x in numeros:
    
    quadrados[x] = x**2
    
print(quadrados)
```

```python
"""
Dicionários e Funções
    Passando dicionários como argumentos para funções
    Retornando dicionários de funções
"""

# 1. Suponha que temos um dicionário representando um perfill de
# usuário e queremos exibir este perfil.

# 2. Vamos criar uma função que pode criar um novo perfil de usuário
# com base em informações fornecidas.

# Definição do dicionário do perfil do usuário

# Iniciando um dicionário chamado 'usuário'
usuario = {
    'nome': 'João',
    'idade': 25,
    'email': 'joao@email.com'
}

def exibir_perfil(perfil):
    
    for chave, valor in perfil.items():
        
        print(f'{chave.title()}: {valor}')
    
exibir_perfil(usuario)

# 2. Retornando dicionários de funções

def criar_perfil(nome, idade, email):
    
    return{
        'nome': nome,
        'idade': idade,
        'email': email
    }
    
novo_usuario = criar_perfil('Ana', 30, 'ana@email.com')

print()

exibir_perfil(novo_usuario)
```

```python
"""
Exercício: Dicionários e Funções

Objetivo: Praticar a interação entre dicionários e funções, aprendendo
a passar dicionários como argumentos e a retornar dicionários a partir
de funções.

Contexto:

Imagine que você está construindo um sistema para uma biblioteca. Esse
sistema deve ser capaz de:

    Registrar um novo livro, fornecendo título, autor e ano de publicação.
    Exibir detalhes de um livro registrado.
    
Instruções:

    1. Registro do Livro:
    
        a) Crie uma função chamada registrar_livro que aceite três
        parâmetros: título, autor e ano.
        
        b) Esta função deve retornar um dicionário representando o livro.
        O dicionário deve ter três chaves: "título", "autor" e "ano", e
        os valores correspondentes devem ser os fornecidos à função.
        
    2. Exibição de livro:
    
        a) Crie uma função chamada exibir_livro que aceite um dicionário 
        (representando um livro) como seu único argumento.
        
        b) A função deve imprimir os detalhes do livro em um formato legível.
        
    Dicas: 
    
        Lembre-se de usar o operador de atribuição = para criar pares
        chave-valor no dicionário.
        
        Ao exibir os detalhes do livro, você pode usar um loop para iterar
        sobre os pares chave-valor do dicionário ou acessar cada chave
        individualmente.
"""

# 1.
def registrar_livro(titulo, autor, ano):
    
    return {
        
        'titulo': titulo,
        'autor': autor,
        'ano': ano
    }
    
livro = registrar_livro('1984', 'George Orwell', 1949)
print(livro)

print()

# 2.
def exibir_livro(livro_dicionario):
    
    print(f'Título: {livro_dicionario['titulo']}')
    
    print(f'Autor: {livro_dicionario['autor']}')
    
    print(f'Ano: {livro_dicionario['ano']}')
    
exibir_livro(livro)
```
